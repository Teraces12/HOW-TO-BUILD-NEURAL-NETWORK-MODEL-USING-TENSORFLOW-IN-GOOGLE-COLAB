- For Example, let's convert 0°C celsius temperature to Fahrenheit:  **(0°C × 9/5) + 32 = 32°F**

# THIRD STEP : **DATASET VISUALIZATION** 

```
plt.scatter(Temp_df['Celsius'], Temp_df['Fahrenheit'],color='r')
```

#FOURTH STEP : **CREATE TESTING AND TRAINING DATASET**

```
X_train = Temp_df['Celsius']
y_train = Temp_df['Fahrenheit']
```

# FIFTH STEP: **BUILD AND TRAIN THE MODEL**

```
X_train.shape
```

```
y_train.shape
```

```
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))

```

```
model.summary()
```

```
model.compile(optimizer=tf.keras.optimizers.Adam(0.5), loss='mean_squared_error')
```

```
epochs_hist = model.fit(X_train, y_train, epochs = 100)
```

```
model.compile(optimizer=tf.keras.optimizers.Adam(1), loss='mean_squared_error')
```

```
epochs_hist = model.fit(X_train, y_train, epochs = 500)
```

# SIXTH STEP: **EVALUATING THE MODEL** 

```
epochs_hist.history.keys()
```

```
plt.plot(epochs_hist.history['loss'],color='r')
```

```
plt.plot(epochs_hist.history['loss'],color='g')
plt.title('Model Loss Progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```
model.get_weights()
```

Use the trained model to perform predictions

```
Temp_C = 0
Temp_F = model.predict([Temp_C])
print('Temperature in degF Using Trained ANN =', Temp_F)
```

```
# Let's confirm with the equation
Temp_F= 9/5*Temp_C + 32
print('Temperature in DegreF using Equation =', Temp_F)
```
